```python
import pandas as pd 
from textblob import TextBlob
```

```python
pd.options.display.max_columns = None
pd.set_option('display.max_colwidth', -1)
```

```python
text_col=['Summary','Space','Description','Neighborhood_Overview','Notes','Transit','Interaction','House_Rules','Host_About','Amenities']
```

```python
df_text = pd.read_csv('listings.csv', sep = ';',usecols= text_col)
```

```python
df_text.shape
```

```python
df_text_subset = df_text[0:100000]
```

```python
df_text_subset.shape
```

```python
df_text_subset.head(3)
```

```python
df_text_subset.isnull().sum()
```

```python
df_text_subset = df_text_subset.fillna("neutral")
```

```python
df_text_subset.isnull().sum()
```

```python
df_text_subset.head(5)
```

```python

for i in range(df_text_subset.shape[0]):## Row
    print(i)
    for j in range(df_text_subset.shape[1]): ## Columns
        print(j)
        temp =df_text_subset.iat[i,j]
        temp = TextBlob(temp)
        sent = 0
        for word in temp.words:
            analysis = TextBlob(word)
            if analysis.sentiment[0]>0:
                sent = sent + analysis.sentiment[0]
            elif analysis.sentiment[0]<0:
                sent = sent + analysis.sentiment[0]
            else:
                sent = sent
        #print('Addition of all polarity \t',sent)
        df_text_subset.iat[i,j] = sent
```
